```python
import pandas as pd
import functools
```

```python
int12 = pd.read_csv("./data/raw-data-nauger-2022-23/12HPE.csv")
int1 = pd.read_csv("./data/raw-data-nauger-2022-23/1DPE.csv")
int3 = pd.read_csv("./data/raw-data-nauger-2022-23/3DPE.csv")
int7 = pd.read_csv("./data/raw-data-nauger-2022-23/7DPE.csv")
int14 = pd.read_csv("./data/raw-data-nauger-2022-23/14DPER.csv")
int14a = pd.read_csv("./data/raw-data-nauger-2022-23/14DPEA.csv")
int14p = pd.read_csv("./data/raw-data-nauger-2022-23/14DPEP.csv")
int21 = pd.read_csv("./data/raw-data-nauger-2022-23/21DPE.csv")
```

```python
int12.head(5)
```

```python
int12_mod = int12.filter(['BlastHitUniprotID', 'Gene Names','log2FoldChange', 'padj', 'Trinity', 'Purpuratus Homolg'], axis=1)
int12_mod.head(5)

int12_mod = int12_mod[int12_mod.padj < 0.05]
int12_mod['padj'].min()
```

```python
int1_mod = int1.filter(['BlastHitUniprotID', 'Gene Names','log2FoldChange', 'padj', 'Trinity', 'Purpuratus Homolg'], axis=1)
int1_mod = int1_mod[int1_mod.padj < 0.05]

int3_mod = int3.filter(['BlastHitUniprotID', 'Gene Names','log2FoldChange', 'padj', 'Trinity', 'Purpuratus Homolg'], axis=1)
int3_mod = int3_mod[int3_mod.padj < 0.05]

int7_mod = int7.filter(['BlastHitUniprotID', 'Gene Names','log2FoldChange', 'padj', 'Trinity', 'Purpuratus Homolg'], axis=1)
int7_mod = int7_mod[int7_mod.padj < 0.05]

int14_mod = int14.filter(['BlastHitUniprotID', 'Gene Names','log2FoldChange', 'padj', 'Trinity', 'Purpuratus Homolg'], axis=1)
int14_mod = int14_mod[int14_mod.padj < 0.05]

int14a_mod = int14a.filter(['BlastHitUniprotID', 'Gene Names','log2FoldChange', 'padj', 'Trinity', 'Purpuratus Homolg'], axis=1)
int14a_mod = int14a_mod[int14a_mod.padj < 0.05]

int14p_mod = int14p.filter(['BlastHitUniprotID', 'Gene Names','log2FoldChange', 'padj', 'Trinity', 'Purpuratus Homolg'], axis=1)
int14p_mod = int14p_mod[int14p_mod.padj < 0.05]

int21_mod = int21.filter(['BlastHitUniprotID', 'Gene Names','log2FoldChange', 'padj', 'Trinity', 'Purpuratus Homolg'], axis=1)
int21_mod = int21_mod[int21_mod.padj < 0.05]
```

```python
dfs = [int12_mod, int1_mod, int3_mod, int7_mod, int14_mod, int14a_mod, int14p_mod, int21_mod]

for i, df in enumerate(dfs, start=1):
    df.rename(columns={col:'{}_int{}'.format(col, i) for col in ('log2FoldChange', 'padj')}, 
              inplace=True)
```

```python
#merge = functools.partial(pd.merge, on=['Trinity', 'BlastHitUniprotID', 'Gene Names', 'Purpuratus Homolg'])
#result = functools.reduce(merge, dfs)
#result.head()
```

```python
result = functools.reduce(lambda  left,right: pd.merge(left,right,on=['Trinity','BlastHitUniprotID', 'Gene Names', 'Purpuratus Homolg'],
                                            how='outer'), dfs).fillna('NaN')
```

```python
list(result.columns.values) #get column names for results df
#for i in dfs:
#    print(len(i.index))

#len(result.index) #get number of row in result df
```

```python
merged_int_data = result[['BlastHitUniprotID',
 'Gene Names',
 'Trinity',
 'Purpuratus Homolg',
 'log2FoldChange_int1',
 'padj_int1',
 'log2FoldChange_int2',
 'padj_int2',
 'log2FoldChange_int3',
 'padj_int3',
 'log2FoldChange_int4',
 'padj_int4',
 'log2FoldChange_int5',
 'padj_int5',
 'log2FoldChange_int6',
 'padj_int6',
 'log2FoldChange_int7',
 'padj_int7',
 'log2FoldChange_int8',
 'padj_int8']]
```

```python
merged_int_data.rename(columns={'log2FoldChange_int1':'log2FC_12hpe',
 'padj_int1':'FDR_12hpe',
 'log2FoldChange_int2':'log2FC_1dpe',
 'padj_int2':'FDR_1dpe',
 'log2FoldChange_int3':'log2FC_3dpe',
 'padj_int3':'FDR_3dpe',
 'log2FoldChange_int4':'log2FC_7dpe',
 'padj_int4':'FDR_7dpe',
 'log2FoldChange_int5':'log2FC_14dpe',
 'padj_int5':'FDR_14dpe',
 'log2FoldChange_int6':'log2FC_14dpea',
 'padj_int6':'FDR_14dpea',
 'log2FoldChange_int7':'log2FC_14dpep',
 'padj_int7':'FDR_14dpep',
 'log2FoldChange_int8':'log2FC_21dpe',
 'padj_int8':'FDR_21dpe'}, inplace=True)
```
